## Models Integration - Gemini, Groq and OpenAI

### OPEN AI Model Integration

```python
import os
from dotenv import load_dotenv

load_dotenv()
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

### OPENAI Model Integration

```python
from langchain.chat_models import init_chat_model

model= init_chat_model("gpt-4.1")
response = model.invoke("How do parrots talk?")
response
```

```python
from langchain_openai import ChatOpenAI

model = ChatOpenAI(model = "gpt-4.1")
respo = model.invoke("Why do parrots talk?")
```

```python
respo
```

### GOOGLE Model Integration

```python
import os
from langchain.chat_models import init_chat_model

os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")
model = init_chat_model("google_genai:gemini-2.5-flash")
response = model.invoke("why do parrots talk?")
response.content
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

model = ChatGoogleGenerativeAI(model = "gemini-2.5-flash-lite")
response = model.invoke("How do parrots talk?")
response.content
```

### GROQ Model Intergration

```python
import os
from langchain.chat_models import init_chat_model

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")

model = init_chat_model("groq:qwen/qwen3-32b")
response = model.invoke("How do parrots talk?")
```

```python
response.content
```

```python
from langchain_groq import ChatGroq

model = ChatGroq(model ="qwen/qwen3-32b")
resp = model.invoke("Why do parrots talk?")
```
